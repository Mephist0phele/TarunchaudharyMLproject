```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("insurance_claim.csv")
df.head()
```

```python
# suffle the dataset
df1=df.iloc[:,0:11]
df1.head()
```

```python
df_1=pd.get_dummies(df1)
df_1.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
df_1=pca.fit_transform(df_1)
```

```python
df_1_=pd.DataFrame(data=df_1)
df_1_.head()
```

```python
df2=df.iloc[:,11:20]
df2.head()
```

```python
df_2=pd.get_dummies(df2)
df_2.head()
```

```python
df3=df.iloc[:,20:26]
df3.head()
```

```python
df_3=pd.get_dummies(df3)
df_3.head()
```

```python
df_final=pd.concat([df_1_,df_2,df_3],axis="columns")
df_final.head()
```

```python
#treating with outliers
from scipy.stats import zscore
zscore=abs(zscore(df_final))
print(df_final.shape)
df_final1=df_final.loc[(zscore<3).all(axis=1)]
print(df_final1.shape)
```

```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
for col in df_final1.columns:
    if df_final1.skew().loc[col]>0.55:
        df_final1[col]=np.log1p(df_final1[col])
```

```python
df_final1_=df_final1.replace(np.NaN,0)
```

```python
x=df_final1_.drop("Total Claim Amount",axis=1)
```

```python
y=df_final1_["Total Claim Amount"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(30,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=r_state)
    lr=linear_model.LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print("r2_score corresponding to random state:",r_state,"is",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print("max r2_score corresponding to",final_r_state,"is",max_r_score)
```
