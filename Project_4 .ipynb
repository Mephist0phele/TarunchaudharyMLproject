```python
df_bikeshare1=df_bikeshare.drop("dteday",axis=1)
```

```python
from scipy.stats import zscore
zscore=abs(zscore(df_bikeshare1))
print(df_bikeshare1.shape)
df_bikeshare2=df_bikeshare1.loc[(zscore<3).all(axis=1)]
print(df_bikeshare2.shape)
```

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
for col in df_bikeshare2.columns:
    if df_bikeshare2.skew().loc[col]>0.55:
        df_bikeshare2[col]=np.log1p(df_bikeshare2[col])
```

```python
df_bikeshare2.skew()
```

```python
import seaborn as sns
sns.pairplot(df_bikeshare2)
```

```python
x=df_bikeshare2.iloc[:,0:14]
```

```python
x.shape
```

```python
y=df_bikeshare2.iloc[:,-1]
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(30,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=r_state)
    lr=linear_model.LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print("r2_score corresponding to random state:",r_state,"is",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print("max r2_score corresponding to",final_r_state,"is",max_r_score)
```
