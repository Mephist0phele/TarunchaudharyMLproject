```python
df_avocado1=df_avocado.dropna(axis=0)
```

```python
df_avocado1.head()
```

```python
#date is present in the objecttype not the datatype so we have change its type to datatype
import warnings
warnings.filterwarnings('ignore')
df_avocado1["Date"]=pd.to_datetime(df_avocado1['Date'])
df_avocado1["Month"]=df_avocado1['Date'].apply(lambda x:x.month)
df_avocado1["Day"]=df_avocado1["Date"].apply(lambda x:x.day)
```

```python
# drop the date column
df_avocado2=df_avocado1.drop("Date",axis=1)
df_avocado2.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,10))
sns.lineplot(x="Month", y="AveragePrice", hue='type', data=df_avocado2)
plt.show()
```

As we can see we have 54 regions and 2 unique types, so it's going to be easy to to transform the type feature to dummies, but for the region its going to be a bit complex, so I decided to drop the entire column.

I will drop the Date Feature as well because I already have 3 other columns for the Year, Month and Day.

```python
df_avocado3=df_avocado2.drop("region",axis=1)
df_avocado3.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_avocado3["type"]=le.fit_transform(df_avocado3["type"])
df_avocado3.head()
```

```python
import seaborn as sns
sns.pairplot(df_avocado3)
```

```python
import numpy as np
for col in df_avocado3.columns:
    if df_avocado3.skew().loc[col]>0.55:
        df_avocado3[col]=np.log1p(df_avocado3[col])
```

```python
df_avocado3.skew()
```

```python
x=df_avocado3.drop("AveragePrice",axis=1)
y=df_avocado3["AveragePrice"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=42)
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
y_pred=lr.predict(x_test)
r2_scr=r2_score(y_test,y_pred)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
pred=dtr.predict(x_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rdr = RandomForestRegressor()
rdr.fit(x_train,y_train)
pred=rdr.predict(x_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

Well as we can see the RMSE is lower than the two previous models, so the RandomForest Regressor is the best model in this case.

```python
#saving the model with RandomForestRegressor
x_test,x_train,y_test,y_train=train_test_split(x,y,random_state=42,test_size=20)
rdr = RandomForestRegressor()
rdr.fit(x_train,y_train)
pred=rdr.predict(x_test)
```
