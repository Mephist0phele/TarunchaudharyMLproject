```python
df_9=df_8.drop("native_country",axis=1)
```

```python
df_final=pd.concat([df_1,df_9],axis="columns")
df_final.head()
```

```python
# treating with outliers
from scipy.stats import zscore
zscore=abs(zscore(df_final))
print(df_final.shape)
df_final_=df_final.loc[(zscore<3).all(axis=1)]
print(df_final_.shape)
```

```python
# treating with skewness
import numpy as np
for col in df_final_.columns:
    if df_final_.skew().loc[col]>0.55:
        df_final_[col]=np.log1p(df_final_[col])
```

```python
x=df_final_.drop("income",axis=1)
```

```python
y=df_final_["income"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(30,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=r_state)
    lr=linear_model.LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print("r2_score corresponding to random state:",r_state,"is",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print("max r2_score corresponding to",final_r_state,"is",max_r_score)
```
