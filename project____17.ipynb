```python
# using a split function and using a same 
df4["Arrival_Time"]=df4["Arrival_Time"].str.split(':').str[0]
df4["Arrival_Hour"]=df4["Arrival_Time"].str.split(':').str[0]
df4["Arrival_Minute"]=df4["Arrival_Time"].str.split(':').str[1]
df4["Arrival_Hour"]=df4["Arrival_Time"].astype(int)
df4["Arrival_Minute"]=df4["Arrival_Time"].astype(int)
```

```python
df4["Dep_Hour"]=df4["Dep_Time"].str.split(":").str[0]
df4["Dep_Minute"]=df4["Dep_Time"].str.split(":").str[1]
df4["Dep_Hour"]=df4["Dep_Hour"].astype(int)
df4["Dep_Minute"]=df4["Dep_Minute"].astype(int)
```

```python
df4['Route_1']=df4['Route'].str.split("→").str[0]
df4['Route_2']=df4['Route'].str.split("→").str[1]
df4['Route_3']=df4['Route'].str.split("→").str[2]
df4['Route_4']=df4['Route'].str.split("→").str[3]
df4['Route_5']=df4['Route'].str.split("→").str[4]
```

```python
df4['Route_1'].fillna("None",inplace=True)
df4['Route_2'].fillna("None",inplace=True)
df4['Route_3'].fillna("None",inplace=True)
df4['Route_4'].fillna("None",inplace=True)
df4['Route_5'].fillna("None",inplace=True)
```

```python
# treating with skewness
import numpy as np
for col in df9.columns:
    if df9.skew().loc[col]>0.55:
        df9[col]=np.log1p(df9[col])
```

```python
# treating with outliers
from scipy.stats import zscore
zscore=abs(zscore(df9))
print(df9.shape)
df9_=df9.loc[(zscore<3).all(axis=1)]
print(df9_.shape)
```

```python
x=df9_.drop("Price",axis=1)
```

```python
y=df9_["Price"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(30,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=r_state)
    lr=linear_model.LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print("r2_score corresponding to random state:",r_state,"is",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print("max r2_score corresponding to",final_r_state,"is",max_r_score)
```

```python
# using a Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=84)
rdf = RandomForestRegressor()
rdf.fit(x_train,y_train)
y_pred=rdf.predict(x_test)
r2_scr=r2_score(y_test,y_pred)
print("r2score:",r2_scr)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test,y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
