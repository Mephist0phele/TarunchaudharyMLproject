```python
#treating the ouliers
from scipy.stats import zscore
zscore=abs(zscore(df1))
print(df1.shape)
df2=df1.loc[(zscore<3).all(axis=1)]
print(df2.shape)
```

```python
# treating with skewed data
for col in df1.columns:
    if df2.skew().loc[col]>0.55:
        df2[col]=np.log1p(df2[col])
```

```python
x=df2.drop("1",axis=1)
```

```python
x.shape
```

```python
y=df2["1"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=35,test_size=20)
                                              
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import  DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import accuracy_score
model= [  SVC(), DecisionTreeClassifier(), KNeighborsClassifier(),
        RandomForestClassifier()]
```

```python
import warnings
warnings.filterwarnings("ignore")
for m in model:
    
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('accuracy score of m',m,'is')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
```

```python

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
print("RMSE is :",np.sqrt(mean_squared_error(y_test,predm)))
```

root mean square is low means model is learning very well
