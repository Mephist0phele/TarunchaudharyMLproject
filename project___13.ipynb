```python
# use label encoder for 
import warnings
warnings.filterwarnings("ignore")
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_housing["ocean_proximity"]=le.fit_transform(df_housing["ocean_proximity"])
df_housing.head()
```

```python
from scipy.stats import zscore
zscore=abs(zscore(df_housing))
print(df_housing.shape)
df_housing=df_housing.loc[(zscore<3).all(axis=1)]
print(df_housing.shape)
```

```python
import seaborn as sns
sns.pairplot(df_housing)
```

```python
df_housing.skew()
```

```python
import numpy as np
for col in df_housing.columns:
    if df_housing.skew().loc[col]>0.55:
        df_housing[col]=np.log1p(df_housing[col])
```

```python
x=df_housing.drop("median_house_value",axis=1)
y=df_housing["median_house_value"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
max_r_score=0
for r_state in range(30,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=r_state)
    lr=linear_model.LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print("r2_score corresponding to random state:",r_state,"is",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print("max r2_score corresponding to",final_r_state,"is",max_r_score)
```
